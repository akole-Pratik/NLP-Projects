# Stemming & Lemmatization
# stemming- Process of reducing the words in its root form. Which are created by removing suffix and prefix used in it.

# Lemmatization- is another technic which is used to reduce word to its normalize form.

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
nltk.download("wordnet")
```

```python
ps = PorterStemmer()
```

```python
mesg1 = ["accepting", "acceptence", "accept", "acceptation"]

for word in mesg1:
  print(ps.stem(word))
```

```python
mesg2 = ["computing", "computation", "compute", "computer"]

for word in mesg2:
  print(ps.stem(word))
```

```python
lemma = WordNetLemmatizer()
```

```python
mesg1
```

# **Summerization**

```python
import nltk

from nltk.tokenize import word_tokenize
nltk.download("punkt")

from nltk.corpus import stopwords
nltk.download("stopwords")

from nltk.stem import WordNetLemmatizer
nltk.download("wordnet")
```

```python
def cleantext(text):
  #step1 -: lower & tokenization
  token = word_tokenize(text.lower())

  #step2 -: filterout all non alphabets
  ftoken = [t for t in token if(t.isalpha())]

  #step3 -: stopwords removal
  stopword = stopwords.words("english")
  ctoken = [t for t in ftoken if(t not in stopword)]

  #step4 -: lemmatization
  lemma = WordNetLemmatizer()
  ltoken = [lemma.lemmatize(t) for t in ctoken]

  #step5 -: joining the lemma token
  return " ".join(ltoken)
```

```python
mesg ="Free...! Free...! Free...! buy 1 get 1 Free...! offer valid in 30 december 2021, hurry up...!"
```

```python
cleantext(mesg)
```

## **CountVectorizer**

```python
mesg = [
        "Hey, let's go to the gym today...!",
        "hey...!, call me @6 in the evening tomorrow...!",
        "Want to go for a walk with my dog...!"
]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
mesg = cv.fit_transform(mesg).toarray()
```

```python
mesg
```

```python
print(cv.get_feature_names())
```

```python
import pandas as pd
df = pd.DataFrame(data=mesg, columns=cv.get_feature_names())
df
```
